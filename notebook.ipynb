```python
df = pd.read_csv("datos_sociales_espana.csv")
```

```python
df.head()
```

```python
df_unemployment = df[df["indicador"] == "Tasa de desempleo"].copy()
```

```python
df_unemployment
```

```python
orden_periodos = ["Nov 2023", "Q3 2025", "Q4 2025"]
```

```python
df_unemployment["periodo"] = pd.Categorical(
    df_unemployment["periodo"],
    categories=orden_periodos,
    ordered=True
)
```

```python
df_unemployment = df_unemployment.sort_values("periodo")
```

```python
df_unemployment[["periodo", "valor", "unidad"]]
```

```python
spec_unemployment = {
    "plot": [
        {
            "mark": "dot",
            "data": {"from": "unemployment"},
            "x": "valor",
            "y": "periodo"
        }
    ],
    "width": 700,
    "height": 220,
    "grid": True,
    "xLabel": "Tasa de desempleo (%)",
    "yLabel": ""
}
```

```python
widget_unemployment = MosaicWidget(
    spec=spec_unemployment,
    data={"unemployment": df_unemployment}
)
```

```python
widget_unemployment
```

```python
df_unemployment["valor_label"] = df_unemployment["valor"].map(lambda x: f"{x:.2f}%")
df_unemployment
```

```python
df_unemployment["color_punto"] = ["#b8c4d6", "#5b84b1", "#1f4e79"]
df_unemployment
```

```python
spec_unemployment_line = {
    "plot": [
        {
            "mark": "lineY",
            "data": {"from": "unemployment"},
            "x": "periodo",
            "y": "valor",
            "stroke": "#5b84b1",
            "strokeWidth": 2,
            "curve": "monotone-x",
            "tip": True
        },
        {
            "mark": "dot",
            "data": {"from": "unemployment"},
            "x": "periodo",
            "y": "valor",
            "fill": "color_punto",
            "r": 5,
            "tip": True
        },
        {
            "mark": "text",
            "data": {"from": "unemployment"},
            "x": "periodo",
            "y": "valor",
            "text": "valor_label",
            "dy": -12,
            "lineAnchor": "bottom",
            "fontSize": 12,
            "fill": "#2a2a2a"
        }
    ],
    "width": 720,
    "height": 360,
    "grid": True,
    "xLabel": "",
    "yLabel": "Tasa de desempleo (%)"
}
```

```python
widget_unemployment_line = MosaicWidget(
    spec=spec_unemployment_line,
    data={"unemployment": df_unemployment}
)
```

```python
df_unemployment["title"] = (
    "Periodo: " + df_unemployment["periodo"].astype(str) +
    "\nValor: " + df_unemployment["valor_label"]
)
df_unemployment[["periodo", "valor_label", "title"]]
```

```python
df_migration = df[df["indicador"] == "Llegadas irregulares"].copy()
```

```python
df_migration
```

```python
orden_migration = ["2024", "2025", "1 ene-30 nov 2025"]
```

```python
df_migration["periodo"] = pd.Categorical(
    df_migration["periodo"],
    categories=orden_migration,
    ordered=True
)
```

```python
df_migration = df_migration.sort_values("periodo")
```

```python
df_migration[["periodo", "valor", "unidad"]]
```

```python
df_migration["valor_label"] = df_migration["valor"].map(lambda x: f"{int(x):,}".replace(",", "."))
df_migration
```

```python
df_migration["color_barra"] = ["#d8dee9", "#88c0d0", "#2e86ab"]
df_migration
```

```python
df_migration["title"] = (
    "Periodo: " + df_migration["periodo"].astype(str) +
    "\nLlegadas: " + df_migration["valor_label"]
)
df_migration[["periodo", "valor_label", "title"]]
```

```python
spec_migration = {
    "plot": [
        {
            "mark": "barY",
            "data": {"from": "migration"},
            "x": "periodo",
            "y": "valor",
            "fill": "color_barra",
            "title": "title",
            "tip": True
        },
        {
            "mark": "text",
            "data": {"from": "migration"},
            "x": "periodo",
            "y": "valor",
            "text": "valor_label",
            "dy": -12,
            "fontSize": 12,
            "fill": "#2a2a2a"
        }
    ],
    "width": 720,
    "height": 380,
    "grid": True,
    "xLabel": "",
    "yLabel": "Llegadas irregulares"
}
```

```python
widget_migration = MosaicWidget(
    spec=spec_migration,
    data={"migration": df_migration}
 )
```
